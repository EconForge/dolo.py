# A Friedmanesque Stochastic Growth Model

Friedman (1957) proposed that consumers understand that some components of income are transitory and others are permanent.  

This turns out to be a pretty good description of both micro and macro income dynamics, and a particularly clear way to think about the dynamics of income and consumption.

Consider an economy with the following features:

1. There is a permanent component of aggregate labor productivity that grows by a stochastic factor $\Psi$ and is subject to mean one IID shocks:
\begin{eqnarray}
P_{t+1} & = & \Psi_{t+1} P_{t}
\end{eqnarray}
1. "Efficiency units" of labor $N P \Theta$ is the Number of hours worked $N$ multiplied by productivity per hour $P$ multiplied by a transitory shock $\Theta$
1. The balance between sources and uses of resources is 
\begin{eqnarray}
\overbrace{C_{t}+K_{t+1}}^{\text{uses}} & = & \underbrace{K_{t}^{\alpha}(\overbrace{\Theta_{t} N_{t} P_{t}}^{L_{t}})^{1-\alpha}}_{\text{production}} + \underbrace{(1-\delta) K_{t}}_{\text{capital after production}} 
\end{eqnarray}
1. Gross output is generated by a Cobb-Douglas production function, where the unconsumed assets from the previous period constitute the current period's capital stock; we assume that the process of production is what causes depreciation so that the capital remaining after production (which could in principle be consumed) is $(1-\delta)K_{t}$.  Rearranging the prior equation, next period's capital is therefore
\begin{eqnarray}
K_{t+1} & = & K_{t}^{\alpha}L_{t}^{1-\alpha} + (1-\delta) K_{t} - C_{t}
\end{eqnarray}

Assuming no population growth, normalizing hours to $N=1$, and assuming CRRA utility $u(C)=(1-\rho)^{-1}C^{1-\rho}$, the representative consumer's problem is therefore to
\begin{eqnarray}
V_{t}(K_{t},P_{t}) & = & \max_{C_{t}}~\left(\frac{C_{t}^{1-\rho}}{1-\rho}\right) + \beta \mathbb{E}_{t}[V_{t+1}(K_{t+1},P_{t+1})]
\\ & \text{s.t.} & \\
K_{t+1} & = & \underbrace{K_{t}^{\alpha}(\Theta_{t} P_{t} )^{1-\alpha} + (1-\delta) K_{t}}_{\equiv M_{t}} - C_{t}
\end{eqnarray}

Now consider the related problem 
\begin{eqnarray}
v_{t}(k_{t}) & = & \max_{c_{t}}~\left(\frac{c_{t}^{1-\rho}}{1-\rho}\right) + \beta \mathbb{E}_{t}[\Psi^{1-\rho}_{t+1}v_{t+1}(k_{t+1})]
\\ & \text{s.t.} & \\
k_{t+1} & = & \left(k_{t}^{\alpha}\Theta_{t}^{1-\alpha} + (1-\delta) k_{t} - c_{t}\right)\Psi_{t+1}^{-1}
\end{eqnarray}

We can show (never mind the algebra) that $V_{t}(K_{t},P_{t}) = P_{t}^{1-\rho}v_{t}(\overbrace{K_{t}/P_{t}}^{k_{t}})$ where the previous problem's $K$ normalized by $P$ becomes lower-case $k$ above.  So we need only to solve the easier problem with a single state variable $k_{t}$ in order to recover the 
solution to the full problem with both $K$ and $P$ as states.  Defining the absolute return on capital as
\begin{eqnarray}
R_{t} & = & \left(\frac{d Y_{t}}{d K_{t}}\right)K_{t}^{\alpha}(P_{t}\Theta_{t})^{1-\alpha}
\\ & = & \alpha (k_{t}P_{t})^{\alpha-1}L_{t}^{1-\alpha}
\\ & = & P_{t} \alpha k_{t}^{\alpha-1}\Theta_{t}^{1-\alpha}
\end{eqnarray}
it is possible to show that the FOC for the single-state problem is 
\begin{eqnarray}
c_{t}^{-\rho} & = & \beta \mathbb{E}_{t}[(R_{t+1}/\Psi_{t+1})(c_{t+1}\Psi_{t+1})^{-\rho}]
\\ 1- \beta \mathbb{E}_{t}[(R_{t+1}/\Psi_{t+1})(c_{t+1}\Psi_{t+1}/c_{t})^{-\rho}] & = & 0
\end{eqnarray}



This means that if we can solve the simpler problem with one state variable, we have also solved the harder one with two states.

Defining end-of-period (normalized) value as a function of end-of-period (normalized) assets $a_{t}=k_{t}^{\alpha}\Theta_{t}^{1-\alpha}-c_{t}$, 
\begin{eqnarray}
\mathfrak{v}_{t}(a_{t}) & = & \mathbb{E}_{t}[\Psi^{1-\rho}v_{t+1}(\Psi_{t+1}^{-1}\left(a_{t}^{\alpha}\Theta_{t+1}^{1-\alpha}-(1-\delta)a_{t}\right)] \\
\mathfrak{v}_{t}^{\prime}(a_{t}) & = & \mathbb{E}_{t}[\Psi^{1-\rho} \Psi_{t+1}^{-1}\overbrace{(\alpha a_{t}^{\alpha-1}\Theta^{1-\alpha}+(1-\delta))}^{\mathsf{R}_{t+1}=d m_{t+1}/dk_{t+1}}v_{t+1}^{\prime}\left(m_{t+1})\right)]
\end{eqnarray}

The Envelope theorem yields the Euler equation according to which $u^{\prime}(c_{t+1})=v^{\prime}(m_{t+1})$ so
\begin{eqnarray}
c_{t}^{-\rho} & = & \beta \mathbb{E}_{t}[\Psi^{-1}\mathsf{R}_{t+1}\Psi^{-\rho}c_{t+1}^{-\rho}] \\
& = & \beta \mathbb{E}_{t}[\Psi_{t+1}^{-1}\mathsf{R}_{t+1}(\Psi_{t+1}c_{t+1})^{-\rho}]
\end{eqnarray}
and the endogenous gridpoints method uses the fact that
\begin{eqnarray}
c_{t} & = & \left(\beta \mathfrak{v}^{\prime}_{t}(a_{t})\right)^{-1/\rho}
\end{eqnarray}
so that if we pick a grid of values of $a_{t,i}$ then from that we can generate the corresponding $c_{t,i}$ and $m_{t,i}=a_{t,i}+c_{t,i}$.

\begin{eqnarray}
   y_{t} & = & K_{t}^{\alpha}(\Theta_{t})^{1-\alpha}(P_{t} L_{t})^{-\alpha} = \Theta_{t}^{1-\alpha} k_{t}^{\alpha}-\delta k_{t}
\\ P_{t+1} & = & P_{t} \Gamma \Psi_{t+1}
\\ K_{t+1} & = & Y_{t} - C_{t}+K_{t}
\\ K_{t+1}/(P_{t+1}L_{t+1}) & = & \left(y_{t}-c_{t}+k_{t}\right)\left(\frac{P_{t}L_{t}}{ P_{t+1} L_{t+1}}\right)
\\ K_{t+1}/(P_{t+1}L_{t+1}) & = & \left(m_{t}-c_{t}\right)\left(\frac{P_{t}L_{t}}{ P_{t+1} L_{t+1}}\right)
\\ k_{t+1} & = & \left(\frac{y_{t}-c_{t}+k_{t}}{  \Gamma \Psi_{t+1}}\right)
\\ k_{t+1} & = & \left(\frac{m_{t}-c_{t}}{  \Gamma \Psi_{t+1}}\right)
\end{eqnarray}
where 
\begin{eqnarray}
\left(\frac{d}{dL_{t}}\right)\left( K_{t}^{\alpha}(\Theta_{t} P_{t} L_{t})^{1-\alpha}-\delta K_{t}\right) & = & (1-\alpha) \Theta_{t} P_{t} (K_{t}^{\alpha}(\Theta_{t} P_{t} L_{t})^{-\alpha}) \\
& = & (1-\alpha) \Theta_{t}^{1-\alpha} P_{t} k_{t}^{\alpha} \\ 
& = & W_{t} 
\end{eqnarray}
so the level of labor income will be $L_{t}W_{t} = (1-\alpha) \Theta_{t}^{1-\alpha} L_{t} P_{t} k_{t}^{\alpha}$
so the $\textit{ratio}$ of effective labor to permanent productivity $P_{t}L_{t}$ is
\begin{eqnarray}
L_{t}W_{t}/P_{t}L_{t} & = & (1-\alpha)\Theta_{t}^{1-\alpha}k_{t}^{\alpha}
\end{eqnarray}

So Euler's theorem holds with respect to gross output for a gross interest factor $\mathbf{R}$ and wage rate $\mathbf{W}$:
\begin{eqnarray}
k_{t+1}^{\alpha}\Theta^{1-\alpha}_{t+1} & = & \overbrace{\left(\alpha k_{t+1}^{\alpha-1}\Theta_{t+1}^{1-\alpha}\right)}^{f'(k)} k_{t+1} + \overbrace{(1-\alpha)\Theta_{t+1}^{1-\alpha}k_{t+1}^{\alpha}}^{L_{t+1}\left[(P_{t+1}L_{t+1})^{-1}dF_{t+1}/dL_{t+1}\right]}
\\ & = & \mathbf{R}_{t+1}k_{t+1}+\mathbf{W}
\end{eqnarray}
\begin{eqnarray}
   \mathbf{R}_{t+1} & = & \left(\frac{\alpha k_{t+1}^{\alpha-1}\Theta_{t+1}^{1-\alpha}}{  \Gamma \Psi_{t+1}}\right)
\end{eqnarray}

Now define end-of-period assets $a_{t} = y_{t}+k_{t}-c_{t}$, and the net interest factor can be defined as 
\begin{eqnarray}
\mathsf{R}_{t+1} & \equiv & \left(\frac{d y_{t+1}}{d a_{t}}\right)  \\ 
 & = & \left(\frac{d y_{t+1}}{d k_{t+1}}\frac{d k_{t+1}}{d a_{t}}\right)  \\ 
& = & \left(\alpha k_{t+1}^{\alpha-1} \Theta_{t+1}^{1-\alpha}-\delta\right)\left(\frac{1}{  \Gamma \Psi_{t+1}}\right)
\end{eqnarray}

Now define end-of-period value 'gothic v' as a function of end-of-period assets:
\begin{eqnarray}
   \mathfrak{v}_{t}(a_{t}) & = & \mathbb{E}_{t}[v_{t+1}]
\end{eqnarray}

\begin{eqnarray}
   v_{t}(m_{t}) & = & \max_{c_{t}} \left(\frac{c_{t}^{1-\rho}}{1-\rho}\right) + \beta \mathfrak{v}_{t}(a_{t})
\end{eqnarray}

FOC:
\begin{eqnarray}
   c_{t}^{-\rho} & = & \beta \mathfrak{v}_{t}^{\prime}(a_{t})\\ 
   c_{t} & = & \left(\beta \mathfrak{v}_{t}^{\prime}(a_{t})\right)^{-1/\rho}
\end{eqnarray}
